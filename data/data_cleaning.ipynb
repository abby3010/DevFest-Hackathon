```python
import pandas as pd
```

```python

ebola_14 = pd.read_csv('ebola/ebola_2014_2016_clean.csv')
ebola_14 = ebola_14[['Date', 'Country', 
                     'Cumulative no. of confirmed, probable and suspected cases',
                     'Cumulative no. of confirmed, probable and suspected deaths']]

ebola_14.columns = ['Date', 'Country', 'Cases', 'Deaths']

# group by date and country
ebola_14 = ebola_14.groupby(['Country'])['Cases', 'Deaths']
ebola_14 = ebola_14.sum().reset_index()
# filling missing values
ebola_14['Cases'] = ebola_14['Cases'].fillna(0)
ebola_14['Deaths'] = ebola_14['Deaths'].fillna(0)
ebola_14['Cases'] = ebola_14['Cases'].astype(int)
ebola_14['Deaths'] = ebola_14['Deaths'].astype(int)
ebola_14
ebola_14 = ebola_14.to_json(orient = 'records')

# x = json.loads(x)
# print(type(x))

with open("outputs/ebola.json", "w") as outfile:
    outfile.write(ebola_14)
```

```python
cholera = pd.read_csv('cholera/data.csv')
cholera = cholera[['Country', 'Year', 'Number of reported cases of cholera',
                   'Number of reported deaths from cholera', 'Cholera case fatality rate', 'WHO Region']]
cholera.columns = ['Country', 'Year', 'Cases',
                   'Deaths', "Fatality rate", "WHO Region"]


cholera
```

```python
# filling missing values
cholera['Cases'] = cholera['Cases'].fillna(0)
cholera['Deaths'] = cholera['Deaths'].fillna(0)
cholera['Fatality rate'] = cholera['Deaths'].fillna(0)

# converting datatypes
for case in range(len(cholera["Cases"])):
    try:
        cholera.at[case, 'Cases'] = int(cholera["Cases"][case])
    except:
        cholera.at[case, 'Cases'] = 0

# converting datatypes
for case in range(len(cholera["Deaths"])):
    try:
        cholera.at[case, 'Deaths'] = int(cholera["Deaths"][case])
    except:
        cholera.at[case, 'Deaths'] = 0

# cholera.info()


x = cholera.groupby('Country')['Cases', 'Deaths'].sum()
x = x.reset_index()

x = x.to_json(orient = 'records')

# x = json.loads(x)
# print(type(x))

with open("outputs/cholera.json", "w") as outfile:
    outfile.write(x)
```

```python
h1n1 = pd.read_csv('h1n1/data.csv')

h1n1.columns = ['Date', 'Country', 'Cases',
                   'Deaths', "Link"]
h1n1 = h1n1.groupby('Country')['Cases', 'Deaths'].sum().reset_index()
h1n1 = h1n1.to_json(orient = 'records')

with open("outputs/h1n1.json", "w") as outfile:
    outfile.write(h1n1)
```

```python
malaria = pd.read_csv('malaria/reported_numbers.csv')
malaria.columns = ['Country', 'Year','Cases', 'Deaths', 'WHO Region']
malaria = malaria.groupby('Country')['Cases', 'Deaths'].sum().reset_index()

malaria = malaria.to_json(orient = 'records')
with open("outputs/malaria.json", "w") as outfile:
    outfile.write(malaria)
```
